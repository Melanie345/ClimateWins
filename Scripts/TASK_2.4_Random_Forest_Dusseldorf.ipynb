### Data Wrangling

```python
#Reducing the data to ONLY DUSSELDORF
Dusseldorf_list = list(x for x in X.columns if x.find('DUSSELDORF') >=0)
Dusseldorf_list
```

```python
# Now let's create a dataframe with those columns
Dusseldorf_df = X[Dusseldorf_list]
Dusseldorf_df.head()
```

```python
Dusseldorf_df.shape
```

```python
# Next we need to reduce the pleasant_weather dataset to ONLY DUSSELDORF
pleasant_weather_Dusseldorf = pleasant_weather['DUSSELDORF_pleasant_weather']
pleasant_weather_Dusseldorf
```

### Reshaping for modeling

```python
X2 = Dusseldorf_df
```

```python
y2 = pleasant_weather_Dusseldorf
```

```python
# Let's turn them from a dataframe into an array
X = np.array(X2)
y = np.array(y2)
```

```python
X.shape
```

```python
y.shape
```

### Data Split

```python
# Split data into test and train sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### Hyperparameter Optimization

```python
# Creating the RF classifier for GRID SEARCH
clf = RandomForestClassifier()
```

```python
# Grid search cv
grid_space = {
    'max_depth': [3, 5, 10, None],
    'n_estimators': [10, 100, 200],
    'max_features': [1, 3, 5, 7],
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [1, 2, 3]  # Updated here - there was a large error showing when I used the recommended '1,2,3'... min_samples_split=1 is not valid because it would imply splitting a node even when there is only one sample
}
```

```python
start = time.time()
grid = GridSearchCV(clf,param_grid=grid_space,cv=3,scoring='accuracy', verbose=3, n_jobs=-1)
model_grid = grid.fit(X_train, y_train)
print('Search took %s minutes' % ((time.time() - start)/60))
```

```python
# Print grid search results
print('Best GRID search hyperparameters are: '+str(model_grid.best_params_))
print('Best GRID search score is: '+str(model_grid.best_score_))
```

### Random Search

```python
# Define the random search cv
rs_space={'max_depth':list(np.arange(10, 100, step=10)) + [None],
              'n_estimators':np.arange(10, 500, step=50),
              'max_features':randint(1,7),
              'criterion':['gini','entropy'],
              'min_samples_leaf':randint(1,4),
              'min_samples_split':np.arange(2, 10, step=2)
         }
```

```python
# Create a RF Classifier
clf2 = RandomForestClassifier()
```

```python
start = time.time()
rf_random = RandomizedSearchCV(clf2, rs_space, n_iter=200, scoring='accuracy', verbose=3, n_jobs=-1, cv=3) #500
model_random = rf_random.fit(X_train, y_train)
print('Search took %s minutes' % ((time.time() - start)/60))
```

```python
# Print random search results
print('Best RANDOM search hyperparameters are: '+str(model_random.best_params_))
print('Best RANDOM search score is: '+str(model_random.best_score_))
```

```python
# Grid search results vs. random search results
# Print grid search results
print('Best GRID search hyperparameters are: '+str(model_grid.best_params_))
print('Best GRID search score is: '+str(model_grid.best_score_))

print('Best RANDOM search hyperparameters are: '+str(model_random.best_params_))
print('Best RANDOM search score is: '+str(model_random.best_score_))
```

```python
Both GRID and RANDOM search scores are 1.0!
```

### Random Forest with Optimized Search Parameters

```python
# creating a RF classifier results from above 
clf3 = RandomForestClassifier(n_estimators = 110, max_depth=60, max_features=6, min_samples_leaf=3, min_samples_split=4, criterion = 'gini')  
  
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf3.fit(X_train, y_train)
```

```python
# performing predictions on the test dataset
y_pred = clf3.predict(X_test)
  
# using metrics module for accuracy calculation
print("Model Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

```python
# Plot tree from the random forest - Unpleasant & pleasant weather
fig = plt.figure(figsize = (80, 40))
plot_tree(clf3.estimators_[15], fontsize = 20, feature_names = Dusseldorf_df.columns, class_names = ['unpleasant', 'pleasant'], filled = True);
```
